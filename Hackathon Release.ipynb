```python
import requests
import json
import re
from ipyleaflet import Map, basemaps, basemap_to_tiles,CircleMarker
import ipyleaflet
```

```python
with open("hackwagon_data.json") as f:
    data = json.load(f)
    f.close()
    
address_list = []
for i in data:
    try:
        address_list.append(i['studentDetails']['address'])
    except:
        pass
postal_codes = []
for i in range(len(address_list)):
    x = re.findall(".*(\d{6}).*$", address_list[i])
    if x != []:
        postal_codes.extend(x)
```

```python
url = "https://maps.googleapis.com/maps/api/geocode/json?key=XXXXXXXXXXXXXXXXXXXXXXXXXXX&address="
coord = []
count = 1

for i in postal_codes[:len(postal_codes)//2]:
    count += 1
#     print(str(count) + "...")
    try:
        coord.append(requests.get(url+i).json()['results'][0]['geometry']['location'])
    except:
#         print("error " + str(i))
        pass
print("Done")
```
